```python
from random import *

def showStatus(strikeCount,ballCount,outCount):
    print("STRIKE: ",strikeCount)
    print("BALL: ",ballCount)
    print("OUT: ", outCount)



arr = []
while 1:
    i = randint(1, 9)
    if arr.count(i) > 0:
        continue
    else:
        arr.append(i)
        if len(arr) == 3:
            break
print(arr)

strikeCount = 0
ballCount = 0
outCount = 0
while 1:
    choice = input()
    for i in range(0, 3, 1):
        for j in range(0, 3, 1):
            #print(type(choice[i]))
            #print(type(arr[j]))
            if int(choice[i]) == arr[j]:
                if i == j:
                    #print("strike")
                    strikeCount += 1
                    break
                else:
                    #print("ball")
                    ballCount += 1
                    break

            if j == 2:
                #print("out")
                outCount += 1
                break
    showStatus(strikeCount,ballCount,outCount)
    if strikeCount == 3:
        break
    strikeCount = 0
    ballCount = 0
    outCount = 0
    
```

```python
class Student:
    name=''
    age=0
    status=''
    def __init__(self, name,age,status):
        self.name=name
        self.age=age
        self.status=status

    def chageStatus(self,status):
        self.status = status

student = Student('이현규',23,'학생')
print(student.name)
print(student.status)
student.chageStatus('바보')
print(student.status)

class CloudStudent(Student):
    room = ''
    def __init__(self, name,age, state,room):
        super(CloudStudent, self).__init__(name,age,state)
        self.room = room
    def setRoom(self, room):
        self.room = room
    def printClass(self):
        print(self.room)

cs = CloudStudent('이혀ㄴ ',24,'천재','4532')
cs.printClass()
cs.chageStatus('dlddld')
print(cs.status)
cs.setRoom('205gh')
cs.printClass()
```
